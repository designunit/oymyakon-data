```python
import json
from pathlib import Path
import pandas as pd
```

```python
data = Path('./oymyakon-Cases.geojson').read_text()
geojson = json.loads(data)
```

```python
def export_feature(f):
    name = f['properties']['name']
    return [
        {
            'name': name,
            'users': x['user'],
            'season': x['season'],
            'topic': x['topic'],
        }
        for x in f['properties']['cases']
    ]
```

```python
items = []
for f in geojson['features']:
    items = items + export_feature(f)

df = pd.DataFrame(items)
df
```

```python
df.to_csv('20190817-oymyakon-cases.csv')
```
